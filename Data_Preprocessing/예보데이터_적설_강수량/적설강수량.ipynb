```python
df1=pd.read_csv("강수량.csv",encoding='euc-kr')
df2=pd.read_csv("적설.csv",encoding='euc-kr')
```

```python
df1["year"]=df1["year"].apply(str)
df1["mon"]=df1["mon"].apply(str)
df1["day"]=df1["day"].apply(str)
```

```python
df2["year"]=df2["year"].apply(str)
df2["mon"]=df2["mon"].apply(str)
df2["day"]=df2["day"].apply(str)
```

```python
df1m = df1["mon"].str.zfill(2)
df1d = df1["day"].str.zfill(2)
df1["date"] = df1["year"].map(str) + df1m + df1d
df1_0 = df1.drop(['year','mon','day'],axis = 1)
df11 = df1_0[['date','area','value']]
df11.columns = ['date','area','강수량']
df11
```

```python
df2m = df2["mon"].str.zfill(2)
df2d = df2["day"].str.zfill(2)
df2["date"] = df2["year"].map(str) + df2m + df2d
df2_0 = df2.drop(['year','mon','day'],axis = 1)
df22 = df2_0[['date','area','value']]
df22.columns = ['date','area','적설']
df22
```

```python
df = pd.merge(df11, df22, on=['date','area'])
df
```

```python
df.to_csv('C:\\Users\\user\\Desktop\\강수량적설.csv', index = False, encoding='euc-kr')
```
